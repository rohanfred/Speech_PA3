```python
import os
import torchaudio
import numpy as np
# from speechbrain.pretrained import SpeakerRecognition
from sklearn.metrics import roc_curve
from tqdm import tqdm
import torch

# Initialize model
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
verification = spkrec.to(device)
# verification = SpeakerRecognition.from_hparams(
#     source="speechbrain/spkrec-ecapa-voxceleb",
#     savedir="pretrained_models/spkrec-ecapa"
# )

# Load trials.txt
TRIALS_PATH = "your_data/trials.txt"
AUDIO_BASE = "your_data/wav"

def compute_eer(scores, labels):
    fpr, tpr, _ = roc_curve(labels, scores)
    eer = fpr[np.nanargmin(np.absolute((1 - tpr - fpr)))]
    return eer

# Read and run trials
scores = []
labels = []
count = 0
with open(evalCsvPath, "r") as f:
    for line in tqdm(f.readlines()):
      count += 1
      label, enroll_rel, test_rel = line.strip().split()
      enroll_path = os.path.join(evalAudioPath, enroll_rel)
      test_path = os.path.join(evalAudioPath, test_rel)

      enroll_signal, _ = torchaudio.load(enroll_path)
      test_signal, _ = torchaudio.load(test_path)

      score, _ = verification.verify_files(enroll_path, test_path)
      scores.append(score)
      labels.append(int(label))
      if count == 50:
        break

# Compute EER
eer = compute_eer(scores, labels)
print(f"\n✅ Equal Error Rate (EER): {eer * 100:.2f}%")
```

```python
import torch.nn as nn

class DoRALinear(nn.Module):
    def __init__(self, in_features, out_features, bias=True):
        super().__init__()
        self.linear = nn.Linear(in_features, out_features, bias=bias)
        self.dora_delta = nn.Parameter(torch.zeros_like(self.linear.weight))

    def forward(self, x):
        return nn.functional.linear(x, self.linear.weight + self.dora_delta, self.linear.bias)
```
